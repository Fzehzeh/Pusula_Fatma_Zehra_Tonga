```python
#yas sutnunda aykiri deger var mi kontrol edildi
def detect_outliers_iqr(df, columns):
    outliers_dict = {}

    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        alt_sinir = Q1 - 1.5 * IQR
        ust_sinir = Q3 + 1.5 * IQR

        # Aykırı değerleri seç
        outliers = df[(df[col] < alt_sinir) | (df[col] > ust_sinir)]
        outliers_dict[col] = outliers

        print(f"{col} sütununda {outliers.shape[0]} aykırı değer var.")

    return outliers_dict



numerical_columns = ['Yas']
outliers = detect_outliers_iqr(df, numerical_columns)
```

```python
def analyze_categorical_outliers_simple(df, threshold_percent=1):
    """
    Kategorik değişkenlerdeki nadir kategorileri tespit eder
    Frekansi %1'den az olan kategoriler aykiri kabul edilir
    """
    
    # Kategorik sütunları al
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    
    # HastaNo gibi ID sütunlarını çıkar
    categorical_columns = [col for col in categorical_columns if 'No' not in col and 'ID' not in col.upper()]
    
    outlier_summary = {}
    
    
    print("KATEGORİK AYKIRI DEĞER ANALİZİ")
   
    
    for col in categorical_columns:
        print(f"\n {col.upper()} Sütunu:")
        print("-" * 40)
        
        # Null olmayan değerleri al
        non_null = df[col].dropna()
        total_count = len(non_null)
        
        if total_count == 0:
            print("Bu sütunda veri yok!")
            continue
            
        # Frekans tablosu
        value_counts = non_null.value_counts()
        value_percentages = (value_counts / total_count) * 100
        
        # Nadir kategoriler (aykırılar)
        rare_categories = value_percentages[value_percentages < threshold_percent]
        normal_categories = value_percentages[value_percentages >= threshold_percent]
        
        # Özet bilgiler
        print(f" Toplam benzersiz kategori: {len(value_counts)}")
        print(f" Toplam kayıt: {total_count}")
        print(f" Eksik veri: {df[col].isnull().sum()}")
        
        # Nadir kategoriler
        if len(rare_categories) > 0:
            print(f"\n AYKIRI KATEGORİLER (<%{threshold_percent}):")
            for category, percentage in rare_categories.items():
                count = value_counts[category]
                print(f" '{category}': {count} adet (%{percentage:.2f})")
            
            outlier_summary[col] = {
                'rare_count': len(rare_categories),
                'rare_categories': rare_categories.index.tolist(),
                'rare_percentages': rare_categories.values.tolist()
            }
        else:
            print(f" Nadir kategori yok (tümü >=%{threshold_percent})")
            
        # Normal kategoriler (en yaygın 3)
        print(f"\n EN YAYGIN KATEGORİLER:")
        for category, count in value_counts.head(3).items():
            percentage = (count / total_count) * 100
            print(f"   • '{category}': {count} adet (%{percentage:.1f})")
    
    return outlier_summary

def visualize_categorical_distribution(df, column, figsize=(12, 6)):
    """
    Tek bir kategorik sütunun dagilimlarini görselleştirir
    """
    # Null olmayan değerleri al
    data = df[column].dropna()
    value_counts = data.value_counts()
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    # Bar plot
    if len(value_counts) > 15:
        # Çok kategori varsa ilk 15'ini göster
        top_categories = value_counts.head(15)
        ax1.barh(range(len(top_categories)), top_categories.values)
        ax1.set_yticks(range(len(top_categories)))
        ax1.set_yticklabels(top_categories.index, fontsize=10)
        ax1.set_title(f'{column} - İlk 15 Kategori')
    else:
        ax1.barh(range(len(value_counts)), value_counts.values)
        ax1.set_yticks(range(len(value_counts)))
        ax1.set_yticklabels(value_counts.index, fontsize=10)
        ax1.set_title(f'{column} - Tüm Kategoriler')
    
    ax1.set_xlabel('Frekans')
    ax1.grid(axis='x', alpha=0.3)
    
    # Pie chart (ilk 8 kategori + diğer)
    top_8 = value_counts.head(8)
    if len(value_counts) > 8:
        other_sum = value_counts.iloc[8:].sum()
        plot_data = list(top_8.values) + [other_sum]
        plot_labels = list(top_8.index) + ['Diğer']
    else:
        plot_data = top_8.values
        plot_labels = top_8.index
    
    ax2.pie(plot_data, labels=plot_labels, autopct='%1.1f%%', startangle=90)
    ax2.set_title(f'{column} - Dağılım Oranları')
    
    plt.tight_layout()
    plt.show()
    
    # Özet istatistik yazdır
    total = len(data)
    unique_count = len(value_counts)
    print(f"\n {column} Özet:")
    print(f"   • Toplam kayıt: {total}")
    print(f"   • Benzersiz kategori: {unique_count}")
    print(f"   • En yaygın: '{value_counts.index[0]}' ({value_counts.iloc[0]} adet)")

def detect_problematic_categories(df):
    """
    Kategorik verilerde sorunlu değerleri tespit eder
    """
    categorical_cols = df.select_dtypes(include=['object']).columns
    problems = {}
    
    for col in categorical_cols:
        col_problems = []
        data = df[col].dropna().astype(str)
        
        # 1. Çok kısa değerler (1-2 karakter)
        short_values = data[data.str.len() <= 2]
        if len(short_values) > 0:
            col_problems.append(f"Çok kısa değerler: {short_values.unique()[:5]}")
        
        # 2. Sayı içeren metin değerler
        numeric_pattern = data[data.str.contains(r'\d', regex=True)]
        if len(numeric_pattern) > 0 and col not in ['TedaviSuresi', 'UygulamaSuresi']:
            col_problems.append("Sayı içeren kategorik değerler var")
        
        # 3. Boşluk karakteri sorunları
        space_issues = data[data.str.contains(r'^\s|\s$', regex=True)]
        if len(space_issues) > 0:
            col_problems.append("Başında/sonunda boşluk olan değerler")
        
        if col_problems:
            problems[col] = col_problems
    
    if problems:
        print("\n TESPİT EDİLEN VERİ KALİTE SORUNLARI:")
        print("="*50)
        for col, issues in problems.items():
            print(f"\n{col}:")
            for issue in issues:
                print(f"    {issue}")
    else:
        print(" Veri kalitesi sorunları tespit edilmedi!")
    
    return problems

# Kullanım örneği - basit ve direkt
def run_categorical_analysis(df):
    """
    Kategorik analizi tek fonksiyonla çalıştır
    """
    print(" Kategorik değişken analizi başlıyor...\n")
    
    # 1. Aykırı değer analizi
    outliers = analyze_categorical_outliers_simple(df, threshold_percent=1)
    
    # 2. Veri kalite kontrolü
    problems = detect_problematic_categories(df)
    
    # 3. Önemli sütunlar için görselleştirme
    important_cols = ['Cinsiyet', 'TedaviAdi', 'Bolum']
    for col in important_cols:
        if col in df.columns:
            print(f"\n {col} sütunu görselleştiriliyor...")
            visualize_categorical_distribution(df, col)
    
    return outliers, problems

# Hızlı özet rapor
def categorical_summary_report(df):
    """
    Kategorik değişkenler için özet rapor
    """
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    print(" KATEGORİK DEĞİŞKENLER ÖZET RAPORU")
    print("="*50)
    
    for col in categorical_cols:
        unique_count = df[col].nunique()
        null_count = df[col].isnull().sum()
        null_percent = (null_count / len(df)) * 100
        
        print(f"{col:20} | Benzersiz: {unique_count:3} | Eksik: {null_count:4} ({null_percent:4.1f}%)")
    
    print("="*50)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Tanilar sütununu temizle ve listeye dönüştür
def clean_tanilar(x):
    if pd.isna(x) or x == 'Bilinmiyor':
        return []
    # Görünmez karakterleri temizle, virgülle ayır, boşlukları temizle, küçük harfe çevir
    x = x.replace('\u200b', '')  # görünmez karakterleri temizle
    return [i.strip().lower() for i in x.split(',') if i.strip() != '']

# Tanilar sütununu gruplara ayır
def group_tanilar(x):
    groups = []
    for tanı in x:
        if 'dorsalji' in tanı:
            groups.append('Tanilar_Dorsalji')
        elif 'omuz' in tanı or 'tendinit' in tanı:
            groups.append('Tanilar_Omuz')
        elif 'aşil' in tanı or 'ayak' in tanı:
            groups.append('Tanilar_Asil_Ayak')
        elif 'diz' in tanı or 'menisküs' in tanı or 'gonartroz' in tanı:
            groups.append('Tanilar_Diz')
        elif 'kalça' in tanı or 'koksartroz' in tanı or 'trokanterik bursit' in tanı:
            groups.append('Tanilar_Kalca')
        elif 'parapleji' in tanı or 'tetrapleji' in tanı or 'hemipleji' in tanı:
            groups.append('Tanilar_Parapleji')
        elif 'hipertansiyon' in tanı or 'diyabetes' in tanı or 'kalp' in tanı:
            groups.append('Tanilar_Kardiyovaskuler')
        elif 'spinal' in tanı or 'disk' in tanı:
            groups.append('Tanilar_Spinal')
        elif 'eklem' in tanı or 'artr' in tanı:
            groups.append('Tanilar_Eklem')
        else:
            groups.append('Tanilar_Diger')
    return list(set(groups))  # Tekrarlayanları kaldır

# Listeyi oluştur
df['Tanilar_list'] = df['Tanilar'].apply(clean_tanilar)
df['Tanilar_grouped'] = df['Tanilar_list'].apply(group_tanilar)

# MultiLabelBinarizer uygula ve int tipine çevir
mlb = MultiLabelBinarizer()
tanilar_encoded = mlb.fit_transform(df['Tanilar_grouped'])
tanilar_df = pd.DataFrame(tanilar_encoded, columns=mlb.classes_).astype(int)

# Orijinal DataFrame ile birleştir
df = pd.concat([df, tanilar_df], axis=1)

# Artık geçici ve orijinal sütunları silebilirsin
df.drop(columns=['Tanilar', 'Tanilar_list', 'Tanilar_grouped'], inplace=True)

# Kontrol
print(df.filter(like='Tanilar_').head())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer


# UygulamaYerleri sütununu temizle ve listeye dönüştür
def clean_uygulama(x):
    if pd.isna(x) or x == 'Bilinmiyor':
        return []
    return [i.strip() for i in x.split(',') if i.strip() != '']

# UygulamaYerleri grubunu belirle
def group_uygulama(x):
    grouped = []
    for loc in x:
        loc_lower = loc.lower()
        if 'ayak' in loc_lower:
            grouped.append('Uygulama_Ayak')
        elif 'el' in loc_lower or 'dirsek' in loc_lower:
            grouped.append('Uygulama_El')
        elif 'diz' in loc_lower or 'menisküs' in loc_lower:
            grouped.append('Uygulama_Diz')
        elif 'omuz' in loc_lower:
            grouped.append('Uygulama_Omuz')
        elif 'kalça' in loc_lower or 'koksartroz' in loc_lower:
            grouped.append('Uygulama_Kalca')
        elif 'bel' in loc_lower:
            grouped.append('Uygulama_Bel')
        elif 'sırt' in loc_lower:
            grouped.append('Uygulama_Sırt')
        elif 'yüz' in loc_lower:
            grouped.append('Uygulama_Yuz')
        elif 'tüm vücut' in loc_lower:
            grouped.append('Uygulama_TumVucut')
    return list(set(grouped))  # Aynı grup birden fazla geçiyorsa tekrarları sil

# Listeyi oluştur
df['UygulamaYerleri_list'] = df['UygulamaYerleri'].apply(clean_uygulama)
df['UygulamaYerleri_grouped'] = df['UygulamaYerleri_list'].apply(group_uygulama)

# MultiLabelBinarizer uygula ve int tipine çevir
mlb = MultiLabelBinarizer()
uygulama_encoded = mlb.fit_transform(df['UygulamaYerleri_grouped'])
uygulama_df = pd.DataFrame(uygulama_encoded, columns=mlb.classes_).astype(int)

# Orijinal DataFrame ile birleştir
df = pd.concat([df, uygulama_df], axis=1)

# Artık geçici ve orijinal sütunları silebilirsin
df.drop(columns=['UygulamaYerleri', 'UygulamaYerleri_list', 'UygulamaYerleri_grouped'], inplace=True)

# Kontrol
print(df.filter(like='Uygulama_').head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def run_full_pipeline(df):
    """
    Tüm veri seti için EDA ve görselleştirme pipeline'ı
    """
    print("Pipeline başlatıldı...\n")
    
    # 1. Sayısal sütunları ayır
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    numeric_cols.remove('TedaviSuresiSeans') if 'TedaviSuresiSeans' in numeric_cols else None
    
    print("Sayısal sütunlar:", numeric_cols)
    
    # Sayısal sütunların korelasyonu
    corr_matrix = df[numeric_cols + ['TedaviSuresiSeans']].corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Sayısal Değişkenler Korelasyon Matrisi")
    plt.show()
    
    # Sayısal sütunlar vs hedef
    for col in numeric_cols:
        plt.figure(figsize=(6,4))
        sns.scatterplot(data=df, x=col, y='TedaviSuresiSeans')
        plt.title(f"{col} vs TedaviSuresiSeans")
        plt.show()
    
    # 2. Kategorik ve one-hot encoded sütunları ayır
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    
    # Önemli one-hot encoded sütunlar
    important_ohe_cols = [col for col in df.columns if set(df[col].unique()) <= {0,1} and col != 'TedaviSuresiSeans']
    
    all_categorical = categorical_cols + important_ohe_cols
    
    print("\nKategorik / One-hot encoded sütunlar:", all_categorical)
    
    for col in all_categorical:
        plt.figure(figsize=(8,4))
        if col in important_ohe_cols:
            # 0/1 sütunları için bar plot
            counts = df[col].value_counts()
            sns.barplot(x=counts.index, y=counts.values)
            plt.title(f"{col} dağılımı (0/1)")
        else:
            # Klasik kategorik sütunlar
            counts = df[col].value_counts()
            sns.barplot(y=counts.index, x=counts.values)
            plt.title(f"{col} dağılımı")
        plt.show()
    
    print("\nPipeline tamamlandı.")

# Kullanım
run_full_pipeline(df)
```

```python
import matplotlib.pyplot as plt

# Kronik hastalıklar sütunlarını seç
chronic_cols = ['aritmi','astım','diyabet','diğer','duchenne musküler distrofisi',
                'fascioscapulohumeral distrofi','hipertiroidizm','hipotiroidizm',
                'kalp yetmezliği','limb-girdle musküler distrofi','myastenia gravis','diger']

# Alerji sütunlarını seç
allergy_cols = ['ilac','polen','toz','yiyecek']

# Fonksiyon: 1'lerin sayısını gösteren bar grafiği
def plot_onehot_distribution(df, columns, title):
    counts = df[columns].sum().sort_values(ascending=False)
    plt.figure(figsize=(10,6))
    plt.barh(counts.index, counts.values, color='skyblue')
    plt.xlabel('Hasta Sayısı')
    plt.ylabel('Kategori')
    plt.title(title)
    plt.gca().invert_yaxis()  # En yüksekten başlasın
    plt.show()

# Kronik hastalıklar grafiği
plot_onehot_distribution(df, chronic_cols, 'Kronik Hastalıkların Yaygınlığı')

# Alerjiler grafiği
plot_onehot_distribution(df, allergy_cols, 'Alerjilerin Yaygınlığı')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Kronik hastalıklar ve alerjiler sütunları
chronic_cols = ['aritmi','astım','diyabet','diğer','duchenne musküler distrofisi',
                'fascioscapulohumeral distrofi','hipertiroidizm','hipotiroidizm',
                'kalp yetmezliği','limb-girdle musküler distrofi','myastenia gravis','diger']

allergy_cols = ['ilac','polen','toz','yiyecek']

# Fonksiyon: 1/0 ile tedavi süresini karşılaştır
def plot_treatment_by_condition(df, columns, title):
    for col in columns:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=df[col], y=df['TedaviSuresiSeans'])
        plt.xlabel(f"{col} (0=Yok, 1=Var)")
        plt.ylabel("Tedavi Süresi (Seans)")
        plt.title(f"Tedavi Süresi vs {col} - {title}")
        plt.show()

# Kronik hastalıklar için
plot_treatment_by_condition(df, chronic_cols, "Kronik Hastalıklar")

# Alerjiler için
plot_treatment_by_condition(df, allergy_cols, "Alerjiler")
```
